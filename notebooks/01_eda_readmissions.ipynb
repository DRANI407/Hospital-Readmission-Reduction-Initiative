# Exploratory Data Analysis — Hospital Readmissions

This notebook loads the **model-ready dataset** and summarizes:
- 30-day readmission rate and target distribution
- Demographics and key clinical variables
- Risk factors (e.g., prior utilization, polypharmacy, length of stay)

**Prerequisites:** Run the pipeline once so `readmission_project/data/model_ready_data.csv` exists (`python run_all.py` from repo root).

```python
import pandas as pd
from pathlib import Path

# Path from repo root (run notebook with cwd = repo root)
ROOT = Path('.').resolve()
if not (ROOT / 'readmission_project').exists():
    ROOT = ROOT.parent
DATA_DIR = ROOT / 'readmission_project' / 'data'
path = DATA_DIR / 'model_ready_data.csv'
df = pd.read_csv(path)
print(f"Shape: {df.shape}")
df.head(10)
```

```python
# Target: 30-day readmission rate
rate = df['target'].mean()
print(f"30-day readmission rate: {rate:.2%}")
print(df['target'].value_counts())
```

```python
# Key risk factors (means by readmission status)
agg = df.groupby('target').agg(['mean', 'count']).T
print(agg)
```
