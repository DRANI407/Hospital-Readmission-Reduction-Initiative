# 01 — Exploratory Data Analysis

Load model-ready data from `data/`, summarize readmission rate, and explore risk factors.

```python
import pandas as pd
from pathlib import Path
ROOT = Path('.').resolve()
if not (ROOT / 'data').exists(): ROOT = ROOT.parent
df = pd.read_csv(ROOT / 'data' / 'model_ready_data.csv')
print(df.shape)
df.head()
```

```python
print('30-day readmission rate:', df['target'].mean().round(4))
df['target'].value_counts()
```

```python
df.groupby('target').agg(['mean', 'count']).T
```
